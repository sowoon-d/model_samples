```python
from soynlp.noun import LRNounExtractor_v2
```

```python
import pandas as pd
```

```python
df = pd.read_csv('result/apy_soyspacing.csv', encoding='utf-8')
```

```python
df.head()
```

```python
df['content'] = df['content'].str.replace(pat=r'[^가-힣A-Za-z0-9]', repl= r' ', regex=True)  # replace all special symbols to space
df['content'] = df['content'].str.replace(pat=r'[\n+]', repl= r' ', regex=True)
df['content'] = df['content'].str.replace(pat=r'[\s\s+]', repl= r' ', regex=True)  # replace multiple spaces with a single space
```

```python
%time sents = df['content'].tolist()
```

## NounLMatchTokenizer

```python
%time noun_extractor = LRNounExtractor_v2(verbose=True, extract_compound=True)
%time noun_scores = noun_extractor.train_extract(sents, min_noun_score=0.3, min_noun_frequency=1)
```

```python
noun_scores
```

## NounLMatchTokenizer

```python
from soynlp.tokenizer import NounLMatchTokenizer
```

```python
lmatch_tokenizer = NounLMatchTokenizer(nouns=noun_scores)

# nouns_ = []
# for sent in sents:
#     nouns_.append(lmatch_tokenizer.tokenize(sent))

nouns_ = list(map(lambda x: lmatch_tokenizer.tokenize(x), sents))
```

```python
nouns_
```

```python
df_Soy_NLMT = pd.DataFrame()
```

```python
df_Soy_NLMT['content'] = nouns_
```

```python
df_Soy_NLMT.head()
```

```python
df_Soy_NLMT.to_csv('result/t_Soy_NLMT.csv', encoding='utf-8', index=False)
```

## WordExtractor
## LTokenizer

```python
from soynlp.word import WordExtractor
from soynlp.utils import DoublespaceLineCorpus

corpus = DoublespaceLineCorpus(sents, iter_sent=True)

word_extractor = WordExtractor(
    min_frequency=100, # example
    min_cohesion_forward=0.05,
    min_right_branching_entropy=0.0
)

word_extractor.train(sents)
words = word_extractor.extract()

cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
```

```python
from soynlp.tokenizer import LTokenizer
tokenizer = LTokenizer(scores=cohesion_score)
```

```python
nouns2_ = list(map(lambda x: tokenizer.tokenize(x), sents))
```

```python
df_Soy_LT = pd.DataFrame()
```

```python
df_Soy_LT['content'] = nouns2_
```

```python
df_Soy_LT.head()
```

```python
df_Soy_LT.to_csv('result/t_Soy_LT.csv', encoding='utf-8', index=False)
```
